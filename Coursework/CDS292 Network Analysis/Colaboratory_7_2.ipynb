## 1. Identifying v-shapes

There is a distinction between counting the v-shapes that have node $h$ as its vertex and checking if nodes $i, h, q$ are a v-shape motif. In this problem, use the IMDB data to print all v-shapes with `Martin Scorsese` as the vertex and `Robert De Niro` as one of the nodes. You can print a 3-element tuple with the nodes involved in the v-shape

````
from readlist import readlist
imdb = readlist('/content/drive/MyDrive/ColabNotebooks/CDS292/imdb.pkl',0)
````

````
def vCheck(G, vertex, end1, end2):
    if G.has_edge(vertex, end1) and G.has_edge(vertex, end2):
        return 1
    else:
        return 0
````

````
nodes_list = imdb.nodes()
v_list = list()

for name in nodes_list:
  if vCheck(imdb, 'Martin Scorsese', 'Robert De Niro', name):
      v_list.append(('Robert De Niro', 'Martin Scorsese', name))
````

````
print(v_list)
````

## <font color='#006633'>3. How slow is using `vCheck` compared to `vcount`?</font>

Let's time both methods for counting v-shapes. A useful piece of information is to have the count of v-shapes for **multiple** nodes in a network. To save this, we can create a dictionary `vAir` with the node labels as the keys, and the number of v-shapes with the node as vertex as the values.

````
nods = ['MZW', 'BZL', 'GGW', 'GGG', 'JSR', 'ROW', 'ABI', 'TOU', 'TID', 'MUW', 'MSS', 'IAO', 'SLH', 'TOG', 'HIB', 'PKR', 'BIR', 'BDP', 'BHR', 'KWF', 'ABS', 'LBP', 'MUK', 'SLX', 'VUS', 'CGM', 'CSH', 'KTL', 'ABR', 'ASV', 'MSW', 'IKO', 'OMO', 'OLF', 'WLH', 'SLN', 'JPR', 'ARW', 'NIQ', 'KID', 'WNH', 'WNR', 'OLA', 'MBI', 'BVH', 'ROO', 'BDU', 'ULZ', 'ULG', 'ULO']
````

````
def vCount(G, i):
  k = G.degree(i)
  return (k * (k - 1) /2)
````

````
import time

t0 = time.time()

vCountAir = {}

for v in nods:
  vCountAir[v] = vCount(airports, v)

print(vCheckAir)

tf = time.time()
print(f'Executed in {tf - t0}')
````

````
print(vCountAir)
````

````
import time

t0 = time.time()

vCheckAir = dict()

for v in nods:
  vCheckCnt = 0
  for i in nods:
    for q in nods:
      if i != q:
        vCheckCnt += vCheck(airports,v,i,q)

  vCheckAir[v] = vCheckCnt/2

print(vCheckAir)

tf = time.time()
print(f'Executed in {tf - t0}')
````
