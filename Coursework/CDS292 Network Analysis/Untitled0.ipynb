```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
for i in range(1,7+1):
  for j in range(1,i):
    G.add_edge(i,j)
```

```python
ci_dict = dict()
for node in G.nodes():
  ci_dict[node] = nx.clustering(G, node)
```

```python
c_sum = 0
for value in ci_dict.values():
  c_sum += value
c_avg = c_sum/G.order()
print(c_avg)
```
