```python
def vi(G, h):
    A = nx.adjacency_matrix(G)
    A = A.toarray()

    nodeset = list(G.nodes())
    idxh = nodeset.index(h)

    n = len(nodeset)

    v = 0
    for i in range(n):
        for q in range(i + 1, n):
            v += A[i][idxh] * A[idxh][q]

    return v
```

```python
vmin = 100
vmax = 0

for nod in G.nodes():
  v = vi(G, nod)
  if v > vmax:
    vmax = v
  if v < vmin:
    vmin = v
```

```python
def ringnet(n):
  r_net = nx.Graph()
  r_net.add_edge(1,n)

  for i in range (1,n):
    r_net.add_edge(i,i+1)

  vcnt = 0
  for nod in r_net.nodes():
    vcnt += vi(r_net,nod)

  print('V is', vcnt)

  print('Avg V is', vcnt/n)
```

```python
ringnet(10)
```
