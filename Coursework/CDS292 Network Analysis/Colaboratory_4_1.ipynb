In the slides and the videos for this lesson, you've seen some examples and different ways to create degree histograms. Let's use another example to understand how the degree histogram can be used to extract some properties from a network.

For this example, create a hub and spoke network with $n = 27$. Then, save its degree histogram as `HSHk` and plot it using bars.

````
import networkx as nx
import matplotlib.pyplot as plt

n = 27

#begin the network
G = nx.Graph()

#add notes and links
for i in range(2,n+1):
  G.add_edge(1,i)

#view network
nx.draw_networkx(G)

'''G_degrees = dict(nx.degree(G))

degree_histogram = {}
for node in G.nodes():
    degree_histogram[node] = 0
for node in G_degrees.keys():
    k = G_degrees[node]
    degree_histogram[k] += 1
"
degree_histogram'''
````

````
def Hdeg(G, FlagZeros=False):
    H = {}
    if FlagZeros:
        n = G.number_of_nodes()
        for k in range(n):
            H[k] = 0

    for node in G.nodes():
        k = G.degree(node)
        H[k] = H.get(k, 0) + 1

    return H

HSND = Hdeg(G)
````

````
plt.title("Histogram using bars of HSHk")
plt.xlabel("k")
plt.ylabel("H(k)")
plt.bar(list(HSND.keys()), list(HSND.values()))
````

For this problem you will redefine the variable `HBHk`. It will still be a dictionary, but this time it will store **several** histograms. Every key of this dictionary will be the number of nodes for a hub and spoke network; and the corresponding values will be the histograms (which are also dictionaries).

As an example, `HSHk[27]` stores the results from the previous problem. In other words `HSHk[n]` $= H(k | n), 0 \leq k \leq n - 1$

Use the following values for $n$: 27, 227, 2227, 22227. Create four plots, one for each Histogram. Use dots and discuss what happens with $H(k = 1)$

**Hint**: for this problem, it's convenient to use a function to produce a hub and spoke network. Also, a `for` loop will help if you iterate through the different values of $n$.

````
def Hdeg(G, FlagZeros=False):
    H = {}
    if FlagZeros:
        n = G.number_of_nodes()
        for k in range(n):
            H[k] = 0

    for node in G.nodes():
        k = G.degree(node)
        H[k] = H.get(k, 0) + 1

    return H

#create hub and spoke network
def create_hsnx(n):
  '''
  input: n (integer) number of nodes
  output: Hub and spoke network with n nodes
  create hub and spoke network
  '''
  hsnx = nx.Graph()
  for i in range(2,n+1):
    hsnx.add_edge(1,i)

  HSHk = Hdeg(hsnx)

  return HSHk
````
