## <font color='#006633'> 1. Line networks </font>

This problem uses a line network, which is a network model you shold know at this point in the semester.

We know there are different ways to create line networks in `Python`, so for this particular problem, you will have to create one with  $n = 15$ using a `for` loop. Then, print the diagram of the resulting network.

````
import networkx as nx

nx.__version__

n = 15

G = nx.Graph()

for i in range(1,n):
  G.add_edge(i,i+1)

nx.draw_networkx(G)
G.nodes()
````

## <font color='#006633'>4. Identifying nodes </font>

The nodes in the airports network are labeled after each three letter code. However, for the purpose of an adjacency matrix there are no labels! Just 0s and 1s. How do we know which row/column represents each one of the nodes?

We could use a dictionary. This problem requires you to create such a dictionary, and then use it. Specifically,

1. Create a dictionary `nodes_id`. The keys are the labels for the nodes in the airports network, and the values are the position of every node in the list `airports.nodes()`. Note that we use that list, because the elements in the adjacency matrix folow the same order as they do in the node set.
2. From the adjacency matrix produced in the previous problem, identify if there is a link between `IAD` and `SCL`

````
nodes_id = dict()

airports_mat = airports_comp.toarray()
print(airports_mat)
n = 0

labels = list(airports.nodes())

for list_ in airports_mat:
  temp_list = list()
  temp_n = 0
  for i in list_:
    if i == 1:
      if labels[n] != labels[temp_n]:
        temp_list.append(labels[temp_n])
    temp_n += 1
  nodes_id[labels[n]] = temp_list
  n += 1
````

````
for st in nodes_id['IAD']:
  if st == 'SCL':
    print(st)
    print('IAD and SCL are linked')
````
