For this problem, we will use a line network with nodes labeled $1, 2, 3, \dots, n$. Think about all possible paths connecting pairs of nodes in this network. Can you identify the **longest** (without visiting the same node more than once) path? It shouldn't be very hard to identify that nodes 1 and $n$ have the longest path in the network. This problem requires you to write code to show that this is in fact true.

Specifically,

1. Create a line network with $n = 200$. Make sure to start labeling your nodes at 1 and not at 0.
2. Calculate the length of the path connecting nodes 1 and 200. You don't need to do this using `Python`
3. Check if there is a path of the length calculated in the previous step connecting nodes 1 and 200. **There should be a path, since it is a line network**
4. Show that there no other path of that length in the network. A useful piece of information is that in a line network, a path that does not repeat visited nodes of length $\ell$ is represented in the adjacency matrix $\mathbf{A}^{\ell}$ with a 1. If it has a 0, it means there is no path, and if there is a number $ > 1$ it means that there are multiple paths and each of them revisits some nodes.  

```
#create line network
G = nx.Graph()
G.add_edges_from([(i, i+1) for i in range(1,200)])

#calc length from node 1 to node 200
length = 0
for i in [(i, i+1) for i in range(1,200)]:
  length += PathCheck(G,(i))
print(length)

#Check if there is a path of the length calculated in the previous step connecting nodes 1 and 200
A = nx.adjacency_matrix(G)
A = A.toarray()

A199 = APower(A, 199)

for i in range(200):
  if A199[0][i] == 1:
    print('There is a L = 199')

#Show that there no other path of that length in the network
L199count = 0
for i in range(200):
  if A199[0][i] == 1:
    L199count +=1
print('The count of L = 199 is', L199count)
```

## <font color='#006633'>4. Counting lengths in a path</font>

Using the dolphins network, select one of the nodes with the lowest degree and one with the highest. Since we know this is a connected network, it means that there is a path between those two nodes. In this problem, you will find the length of the shortest path that connects these nodes.

To achieve this, once the nodes have been identified, create a loop that systematically increases the value of $\ell$. Take the $\ell$-th power of the adjacency matrix and check if there is a path between the two nodes. Once you find it, break the loop and you will find the smallest value of $\ell$ that contains one (or more) paths between the nodes.

```
import networkx as nx

def PathCheck(G, PotentialPath):
  i = PotentialPath[0]
  for h in PotentialPath[1: ]:
    if not G.has_edge(i, h):
      return 0
    i = h
  return 1

G = nx.Graph()
for node in ['Testing Center','Dorm','Cafetria']:
  G.add_edge(node, 'Lecture Hall')
edges = [('Lecture Hall', 'Testing Center'), ('Lecture Hall', 'Dorm'), ('Lecture Hall', 'Cafetria')]
count = 0

nx.draw(G)

for i,q in edges:
  for j in G.nodes:
    count += PathCheck(G, [i,q,j])

print(count)
```
