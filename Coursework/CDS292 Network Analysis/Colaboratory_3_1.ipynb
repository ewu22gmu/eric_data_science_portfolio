Use the following code to create network `G`. Print **all** link indicators of node `d`

Print the results as a list

````
G = nx.Graph()
G.add_node('a')
G.add_edge('c', 'b')
G.add_edge('d', 'a')
G.add_edge('f', 'a')
G.add_edge('d', 'e')
G.add_edge('e', 'a')
G.add_edge('x', 'z')

nx.draw_networkx(G)

def a(G, i, j):
    if G.has_edge(i, j):
        return 1
    else:
        return 0


KofD = list()

for node in G.nodes():
  if node != 'd':
    KofD.append(a(G, 'd', node))

print(f"The link indicators for node d are {KofD}")
````
