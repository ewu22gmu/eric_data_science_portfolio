## 1. What if? [3 points]
Most of the people involved in the Watergate case never took a class on game theory. Therefore, they are not familiar with the [prisoner's dilemma](https://en.wikipedia.org/wiki/Prisoner%27s_dilemma) and think that their best strategy is **not** to testify against anyone else. `Magruder`, however, is aware that the optimal strategy is to testify and therefore, it's links remain intact.

Create this network based on Watergate. I.e. use the the same nodes, but create a network that only contains the links in which `Magruder`is involved. Draw your resulting network using `networkx`

```
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
#from readlist import readlist

G = nx.Graph()

VG = ['Segretti','Chapin','O\'Brien','Parkinson','Krogh','Mitchell','Porter','Strachan','Hunt(H)','Liddy','Magruder','Nixon','Sturgis','Baldwin','McCord','Dean','Haldeman','Martinez','Colson','Kalmbach','Gray','Ehrlichman','Barker']
EG = [('Magruder','Dean'),('Magruder','Haldeman'),('Magruder','Liddy'),('Magruder','Mitchell'),('Magruder','Nixon'),('Magruder','Porter'),('Magruder','Strachan')]

G.add_nodes_from(VG)

G.add_edges_from(EG)

nx.draw_networkx(G)
```

## 3. Simulate network data [4 points]

Throughout this course, we will learn and practice several network properties. It's always useful to be able to recreate network data. This problem requires you to use `networkx` to produce a network with

- $n = 900$
- $m  = 10000$

We learned from the slides of this lesson that we could obtain the density of the network using:

$$ \rho = \frac{2m}{n(n - 1)} \approx 0.0247$$

Test this quantity using your simulated network with the `density()` method in `networkx`

```
import random

G2 = nx.Graph()
G2.add_nodes_from([i for i in range(1,900+1)])
EG = list()


while len(EG) < 10000:
  randomint1 = random.randint(1,900)
  randomint2 = random.randint(1,900)
  if (randomint1, randomint2) not in EG and (randomint2, randomint1) not in EG:
    EG.append((randomint1, randomint2))



G2.add_edges_from(EG)
```

```
nx.density(G2)
```

```
n = G2.order()
m = G2.size()

print(f"n is: {n} and m is: {m}")
```
