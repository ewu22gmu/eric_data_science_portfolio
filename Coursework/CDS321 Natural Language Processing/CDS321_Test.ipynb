```python
#bi-grams words that need to be taken together for meaning (los angeles)
#co-locations words that are frequently together (red wine, white wine)
from nltk.collocations import *

#get bigrams from state of union address
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(text4)
finder.nbest(bigram_measures.pmi, 10) #find bigrams with a pmi of > 10
```

```python
#find unusual words from a text

def unusual_words(text):
  text_vocab = set(w.lower() for w in text if w.isalpha())
  english_vocab = set(w.lower() for w in nltk.corpus.words.words())
  unusual_words = text_vocab - english_vocab
  return sorted(unusual_words)

unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))[:10]
```

```python
#get stopwords corpus
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords.words('english')
```

```python
#find pct of stop words

def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)

print(text1)
content_fraction(text1)*100 #pct
```

```python
for fileid in gutenberg.fileids():
  num_chars = len(gutenberg.raw(fileid))
  num_words = len(gutenberg.words(fileid))
  num_sents = len(gutenberg.sents(fileid))
  num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))

print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
```

```python
from nltk.corpus import brown #import brown
brown.categories() #get catagories
```

```python
brown.words(categories='news')[:5] #get first 5 words
```

```python
brown.sents(categories=['news','editorial'])[:5] #get first 5 sentances
```

```python
#find counts of modal verbs
from operator import methodcaller
newstext = brown.words(categories='news')
Fdist = nltk.FreqDist(w.lower() for w in newstext)
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
  print(m + ':', Fdist[m], end=' ')
```

```python
#condiional freq dist

cdf = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cdf.tabulate(conditions=genres, samples=modals)
```
