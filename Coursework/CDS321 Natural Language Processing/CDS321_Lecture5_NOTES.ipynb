````python
# pull html file
import urllib.request

def get_text(url):
  response = urllib.request.urlopen(url)
  text = response.read().decode('utf-8')
  text = re.sub('<.*?>','',text)
  text = re.sub('\s+',' ',text)
  return text
````

```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer

# Define the corpus
corpus = [
    'Can you guess my language, and how accurate are you in detecting it',
    'Puedes adivinar mi lenguaje y qu precisin tienes para detectarlo',
    'Kannst du meine Sprache erraten und wie genau erkennst du sie?',
    '        ,         '
]

# Create the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the corpus
tfidf_matrix = vectorizer.fit_transform(corpus)

# Get the feature names (tokens)
feature_names = vectorizer.get_feature_names_out()

# Create a dictionary to map tokens to their TF-IDF weights
tfidf_dict = {}
for i, term in enumerate(feature_names):
    tfidf_dict[term] = tfidf_matrix[0, i]  # Assuming you want the weights for the first document

# Print the TF-IDF weights for all tokens
for term, weight in tfidf_dict.items():
    print(f"{term}: {weight}")

# Print the weight of the "" token
print(f"\nWeight of '' token: {tfidf_dict.get('', 0)}")
```



````python
from sklearn.feature_extraction.text import TfidfVectorizer
import operator
````

````python
str1 = 'Can you guess my language, and how accurate are you in detecting it'
str2 = 'Puedes adivinar mi lenguaje y qu precisin tienes para detectarlo'
str3 = 'Kannst du meine Sprache erraten und wie genau erkennst du sie?'
str4 = '        ,         '
````

````python
cps = [str1, str2, str3, str4]
````

````python
vocab = set()

for doc in cps:
  vocab.update(doc.split())

vocab = list(vocab)

word_index = {w: idx for idx, w in enumerate(vocab)}

tfidf = TfidfVectorizer(vocabulary=vocab,analyzer='word')

tfidf_matrix = tfidf.fit_transform(cps)
````

````python
word_to_check = ""
if word_to_check in word_index:
    word_index_in_vocab = word_index[word_to_check]
    for i, doc in enumerate(cps):  # Iterate through documents
        tfidf_value = tfidf_matrix[i, word_index_in_vocab]
        print(f"TF-IDF for '{word_to_check}' in document {i+1}: {tfidf_value}")
else:
    print(f"Word '{word_to_check}' not found in vocabulary.")
````

````python
df = pd.DataFrame(tfidf_matrix.toarray(), columns=vocab)
df['you']
````
