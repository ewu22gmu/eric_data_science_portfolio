# Task 2
List the top 5 pairs of actors who have been in the most movies together.

```python
#merge over mid to make pairs
actors2x = pd.merge(E.inmoviedf[['mid','pid']],
                    E.inmoviedf[['mid','pid']],
                    on='mid',
                    suffixes=('_1', '_2')
                ) 

#remove self pairs
actor_pairs = actors2x[actors2x['pid_1'] > actors2x['pid_2']].copy() 

#group by pairs, assign counts, reset label, then sort by movie count; only show the top 5 pairs
print(actor_pairs.groupby(['pid_1', 'pid_2']).size().reset_index(name='movie_count').sort_values(by='movie_count', ascending=False).head(5))
```

# Task 3
Most movies have several actors, and they have birth dates. Define _mbd_ as the "movie birth difference" which is the difference (in months) of the ages of the oldest and youngest actors in that movie. Find the 5 movies with the largest _mdb_ values.

You may want to solve this problems in steps. PLEASE explain what you are doing at each step. This can be either as code comments or Jupyter text cells.

```python
#get actors from inmoviedf
actors = pd.merge(E.inmoviedf[['mid','pid']],
                  E.persondf[['pid', 'birth', 'death']],
                  on='pid') #inner join to get birth and death to find age

#add age
actors['age'] = np.where(actors['death'] > 0, actors['death'] - actors['birth'], E.month - actors['birth'])

#find the lowest and highest age per movie
diff_in_age = actors.groupby('mid')['age'].aggregate(['min','max']).reset_index() 
diff_in_age['mbd'] = diff_in_age['max'] - diff_in_age['min'] #calculate mbd

#sort by highest difference and only show top 5
diff_in_age[['mid','mbd']].sort_values(by='mbd', ascending=False).head(5) 
```
