What is the percentage of immigrants who have married someone of a different dominant ethnic group (dometh)? 

1. Create _df1_ which is the merge of _portdf_ and _persondf_. The only columns you need from _persondf_ are [pid, dometh]
1. Create _df2_ which is the merge of _df1_ and _weddf_ where this collects the wedding information if the immigrant is female.
1. Create _df3_ which is the merge of _df2_ and _weddf_ where this collects the wedding information if the immigrant is male. Use __copy()__ in the creation of _df3_ because we will add columns soon.
1. Use __fillna(-1)__ to get rid of the NaN values.
1. Create a new column named _myspouse_ which is the _pid_ value of the immigrant's, spouse.
1. Create _df4_ from the merge of _df3_ and _persondf[pid,dometh]_ to get the dominant ethnicity of the spouse.
1. Count the number of rows in which the dominant ethnicity of the two spouses differs. Divide this by the total number of rows to get the percentage.

```python
#identify inter-ethnicity marriges pass 2

#enrich portdf to include dometh and drop emigdate from df1_1
df1_1 = pd.merge(left=E.portdf,right=E.persondf[['pid','dometh']],on='pid').drop('emigdate',axis=1)

#get husbands pid and wife pid from the weddf
cols_1 = ['husband','wife']
df_hw = E.weddf[cols_1]

#create and enrich df2_1 by adding husbands and drop the wife column since we merged over it
df2_1 = pd.merge(left=df1_1, right=df_hw, left_on='pid', right_on='wife',how='left').drop('wife',axis=1)

#enrich df2_1 by adding wifes and drop the husband_w column since we merged over it
df2_1 = pd.merge(left=df2_1, right=df_hw, left_on='pid', right_on='husband',how='left',suffixes=('','_w')).drop('husband_w',axis=1)
#print(df2_1)

#create myspouse column in df2_1
df2_1['myspouse'] = np.where(df2_1['husband']>0, df2_1['husband'], df2_1['wife'])

#get myspouse dometh and drop pid_myspouse since we merged over it
df3_1 = pd.merge(left=df2_1,right=E.persondf[['pid','dometh']],left_on='myspouse',right_on='pid',suffixes=('','_myspouse')).drop('pid_myspouse',axis=1)


#filter out unmarried imigrants and identify cases of interethnicity marriage
s1 = df3_1['myspouse'] > 0 
s2 = df3_1['dometh'] != df3_1['dometh_myspouse']
query = s1 & s2

#output married & interethnicity marriges/total marriges
print(len(df3_1.loc[query])/len(df3_1.loc[s1])) 
```

```python
#identify inter-ethnicity marriges pass 2
df2_1 = pd.merge(
            left=pd.merge(
                left=pd.merge(left=E.portdf,
                              right=E.persondf[['pid','dometh']],
                              on='pid'
                            ).drop('emigdate',axis=1), 
                        right=E.weddf[['husband','wife']], 
                        left_on='pid', 
                        right_on='wife',
                        how='left'
                    ).drop('wife',axis=1), 
                    right=E.weddf[['husband','wife']],
                    left_on='pid', 
                    right_on='husband',
                    how='left',suffixes=('','_w')
                ).drop('husband_w',axis=1)

#create myspouse column in df2_1
df2_1['myspouse'] = np.where(df2_1['husband']>0, 
                             df2_1['husband'], 
                             df2_1['wife']
                            )

#get myspouse dometh and drop pid_myspouse since we merged over it
df3_1 = pd.merge(left=df2_1,
                 right=E.persondf[['pid','dometh']],
                 left_on='myspouse',
                 right_on='pid',
                 suffixes=('','_myspouse')
                ).drop('pid_myspouse',axis=1)


#filter out unmarried imigrants and identify cases of interethnicity marriage
s1, s2 = df3_1['myspouse'] > 0, df3_1['dometh'] != df3_1['dometh_myspouse']

#output married & interethnicity marriges/total marriges
print(len(df3_1.loc[s1 & s2])/len(df3_1.loc[s1])) 
```

Create _df1_ which is the merge of _portdf_ and _persondf_. The only columns you need from _persondf_ are [pid, dometh]

Create _df2_ which is the merge of _df1_ and _weddf_ where this collects the wedding information if the immigrant is female.

```python
df2 = pd.merge(left=df1,right=E.weddf,left_on='pid',right_on='wife',how='left')
len(set(df2['pid']))
```

Create _df3_ which is the merge of _df2_ and _weddf_ where this collects the wedding information if the immigrant is male. Use __copy()__ in the creation of _df3_ because we will add columns soon.

```python
df3 = pd.merge(left=df2,right=E.weddf,left_on='pid',right_on='husband',how='left').copy()
len(set(df3['pid']))
```

Create a new column named _myspouse_ which is the _pid_ value of the immigrant's, spouse.

```python
xcol = ['cid_x', 'husband_x', 'wife_x', 'firstH_x', 'lastH_x', 'firstW_x', 'lastW_x', 'date_x', 'divorce_x', 'spdie_x']
ycol = ['cid_y', 'husband_y', 'wife_y', 'firstH_y', 'lastH_y', 'firstW_y', 'lastW_y', 'date_y', 'divorce_y', 'spdie_y']
col = list(E.weddf.columns)

for i in range(len(col)):
    df3[col[i]] = df3[xcol[i]].combine_first(df3[ycol[i]])

df3 = df3.drop(columns=xcol+ycol)

df3['myspouse'] = np.where(df3['husband'] != df3['pid'], df3['husband'], df3['wife'])
```

Create _df4_ from the merge of _df3_ and _persondf[pid,dometh]_ to get the dominant ethnicity of the spouse.

```python
df4 = pd.merge(left=df3,right=E.persondf[['pid','dometh']],left_on='myspouse',right_on='pid',how='left',suffixes=('', '_spouse')).fillna(-1)
```

Count the number of rows in which the dominant ethnicity of the two spouses differs. Divide this by the total number of rows to get the percentage.

```python
s1 = df4['myspouse'] > 0 #check marrige
s2 = df4['dometh'] != df4['dometh_spouse'] #check diff dometh
query = s1 & s2

len(df4.loc[query])/len(df4.loc[s1])
```
