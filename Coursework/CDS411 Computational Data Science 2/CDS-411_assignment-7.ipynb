## Problem 1
A object moves with a speed of $v_1 = 20.5$ m/s for $t_1=112$ s, then instantly changes speed to $v_2=17.6$ m/s.  It travels for $t_2 = 423$ s at this new speed.  How far did the car travel for the entire journey?

```python
# your code and numerical answer
v1 = 20.5
t1 = 112
v2 = 17.6
t2 = 423

d1 = v1*t1
d2 = v2*t2
d = d1 + d2

print(f"The car traveled {d} meters.")
```

## Problem 3
An object obeys simple projectile motion.  It is launched with a velocity of $v=20.5$ m/s at an angle of $\theta =26.7^\circ$.  What is the peak height?  What is the horizontal distance traveled?

```python
# your code and numerical answers
v = 20.5; g = -9.81
theta = np.radians(26.7)

vx = v*np.cos(theta)
vy = v*np.sin(theta)

t = np.roots((0.5*g, vy, 0)); t = t[0]

print(t)

mx_height = 0 + vy*t/2 + 0.5*g*(t/2)**2
distance_x = vx*t

print(f"The peak height is {mx_height} meters.")

print(f"The horizontal distance traveled is {distance_x} meters.")
```

## Problem 4
The object is launched from a height $y_0=2$ m, with a velocity of $v=20.5$ m/s at an angle of $\theta =26.7^\circ$.  What is the peak height?  What is the horizontal distance traveled?

```python
# your code and numerical answers

v = 20.5; g = -9.81; y0 = 2
theta = np.radians(26.7)

vx = v*np.cos(theta)
vy = v*np.sin(theta)

t = -vy/g

print(t)

mx_height = y0 + vy*t + 0.5*g*(t)**2
distance_x = (t + np.roots((0.5*g, 0, mx_height))[0])*vx

print(f"The peak height is {mx_height} meters.")

print(f"The horizontal distance traveled is {distance_x} meters.")
```

## Problem 8
Use the leapfrog method.  A spring starts with $k=10$ N/m, $m=2$ kg, and $\Delta x = 0.3$ m.  Each time the mass passes through the center position (in either direction) the velocity is reduced by 10 %.  Create a plot of the $x$ position for every 0.1 seconds for 1000 iterations.

```python
# your code and numerical answers
def Leapfrog(niter=200):
  xx = [] # position
  vv = [] # velocity
  k = 10 # spring constant
  x = 0.3 # initial stretch
  m = 2 # mass
  v1 = 0 # initial velocity
  dt = 0.1 # time step
  x1 = x

  for i in range( niter ):
    a2 = -k * x1 / m
    v2 = v1 + a2 * 0.5*dt
    v2 = v2*0.9 if abs(x1) <= 0.01 else v2
    x2 = x1 + v1 * dt + 0.5 * a2 * dt * dt

    v1 = v2; x1 = x2
    xx.append(x2); vv.append(v2)
  return xx, vv

xx, vv = Leapfrog(1000)
plt.plot(xx, label = 'Position')
plt.plot(vv, label = 'Velocity')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Position/Velocity')
plt.show()
```

## Problem 10
Use the leapfrog method with $k=10$ N/m, $\Delta x=0.3$ m, $m=2$ kg.  Each time the mass passes through the center (either direction) lower the value of $k$ by 5 %.  This spring has limits and it can not go past $x > 0.3$ m or $x < -0.3$ m.  Each time that $x$ is computed to be beyond these limits, set $x$ to the limit (either 0.3 m or -0.3 m) and set $v=0$ m/s.

```python
# your code and numerical answers

def Leapfrog(niter=200):
  xx = [] # position
  vv = [] # velocity
  k = 10 # spring constant
  x = 0.3 # initial stretch
  m = 2 # mass
  v1 = 0 # initial velocity
  dt = 0.1 # time step
  x1 = x

  for i in range( niter ):
    k = k*0.95 if abs(x1) <= 0.03 else k
    a2 = -k * x1 / m
    v2 = v1 + a2 * 0.5*dt
    x2 = x1 + v1 * dt + 0.5 * a2 * dt * dt

    if x2 > 0.3:
      x2 = 0.3; v2 = 0
    elif x2 < -0.3:
      x2 = -0.3; v2 = 0

    v1 = v2; x1 = x2
    xx.append(x2); vv.append(v2)
  return xx, vv

xx, vv = Leapfrog(1000)
print(xx)

plt.plot(xx, label = 'Position')
plt.plot(vv, label = 'Velocity')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Position/Velocity')
plt.show()
```
