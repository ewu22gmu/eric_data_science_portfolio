## Problem 1
A model of exponential population growth described by the equation,

$$\frac{ {\tt d}p}{{\tt d} t} = kp$$

Create a model function for this equation.  Create time sampling points from 0 to 20.  Run __odeint__ using the model function.  Plot the result.

```python
# your function, code, and plot
def model(p,t):
  k = 0.1
  dpdt = k*p
  return dpdt

inits = [5]
ts = np.linspace(0,20,100)
dp = si.odeint(model,inits,ts)

plt.scatter(ts,dp,marker='.')
plt.show()
```

## Problem 4
A heated object will cool by the equation

$$\frac{{\tt d}T}{{\tt d}t} = -k (T-T_c)$$

where $T$ is object temperature, $k$ is a constant, and $T_c$ is the environment temperature (not the object temperature).

Use $k=1$.  The initial object temperature is 212 and the environment temperature is 32.  Model this system and include the chart for time 0 to 20.  Plot the result.

```python
# your function, code, and plot
def model(T,t):
  k = 1
  T_c = 32
  dTdt = -k*(T-T_c)
  return dTdt

inits = [212]
ts = np.linspace(0,20,100)
dT = si.odeint(model,inits,ts)

plt.scatter(ts,dT,marker='.')
plt.show()
```

## Problem 5
Growth of bacteria initially increases, but then saturation occurs.  It is modeled by,

$$\frac{{\tt d}P}{{\tt d} t} = k P \left( 1 - \frac{P}{P_{max}}  \right) $$

where $P$ is the population count, $k=1$, and $P_{max}$ is the maximum population allowed by the environment.

Model this function for time 0 to 20 with an initial population of 1 and a max population is 10,000.

Model this system and show a plot of your result. Plot the result.

```python
# your function, code, and plot
def model(P,t):
  k = 1
  P_max = 10000
  dPdt = k*P*(1-P/P_max)
  return dPdt

inits = [1]
ts = np.linspace(0,20,100)
dP = si.odeint(model,inits,ts)

plt.scatter(ts,dP,marker='.')
plt.show()
```

## Problem 6
Model the following system,
$$ \frac{{\tt d}x}{{\tt d}t } = -y + 1.8z $$

$$ \frac{{\tt d}y}{{\tt d}t } = 2x -2y+2z $$

$$ \frac{{\tt d}z}{{\tt d}t } = -3x -2.5y +0.5z $$

Use the initial conditions $x_0=-0.3$, $y_0=0.4$, and $z_0=0.1$.  The time range is 0 to 10.  Model this function and plot all three variables on a single chart.

```python
# your fucntion, code, and plot
def model(xyz,t):
  x, y, z = xyz
  dxdt = -y + 1.8*z
  dydt = 2*x - 2*y + 2*z
  dzdt = -3*x - 2.5*y + 0.5*z

  answ = (dxdt, dydt, dzdt)
  return answ

inits = (-0.3, 0.4, 0.1)
ts = np.linspace(0,10,100)
xyzs = si.odeint(model,inits,ts)

x = xyzs[:,0]; y = xyzs[:,1]; z = xyzs[:,2]

plt.scatter(ts,x,label='x',marker='.')
plt.scatter(ts,y,label='y',marker='.')
plt.scatter(ts,z,label='z',marker='.')
plt.legend()
plt.show()
```
