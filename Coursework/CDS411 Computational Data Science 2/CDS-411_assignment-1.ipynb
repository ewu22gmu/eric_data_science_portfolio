## Problem
Create a list of 100,000 random integers between 0 and 20.  Create a new list which is the subtraction of consecutive values in the original list.  Example: original list = [1,5,3,5...].  New list [4,-2,2,...]  The length of the new list should be 99,999.  Find the min and max of this list.  Create a histogram which has bins ranging from the min to the max.  The number of bins should be max-min.

```python
og_list, cnsc_list = list(), list() #Initialize vars

for n in range(100000):
  og_list.append(np.random.randint(0, 20)) #Add rand int to og list
  if n != 1:
    cnsc_list.append(og_list[n] - og_list[n-1]) #Add rand int to cnsc list

cnsc_arr = np.array(cnsc_list)
```

```python
print(len(og_list), len(cnsc_arr)) #Sanity check
```

```python
min_num, max_num = np.min(cnsc_arr), np.max(cnsc_arr) #Find min, max

print(f"The min is {min_num}. The max is {max_num}")

plt.hist(cnsc_arr, bins=(max_num - min_num))
plt.plot()
```
