## Problem 1
Given below are the counts for the first five positions of the sentences from "Mr. Raspberry" by H. W. Monroe.  The text was split into its sentences, and the occurrences of the first five letters were counted, but only a few are shown.  The first row of data goes with the first position in the sentences.  This shows that the most popular first letter was actually a double-quote, and it was seen 1162 times.  The second most popular letter was a 't' (328 times).  For the second position in the sentences, the most popular letter was 'h' (539 times) and the second most popular was 'y' (304 times).  Using this data, draw an emission hidden Markov model (HMM) with the probabilities of emission.

```python
### answer

inpt = np.array([[('\"', 1162), ('t', 328), ('s', 199), ('k', 178), ('a', 122)], [('h', 539), ('y', 304), ('i', 209), ('a', 172), ('w', 167)],[('e', 629), ('o', 317), ('h', 299), (' ', 229), ('l', 214)],[(' ', 600), ('i', 308), ('e', 218), ('a', 188), ('n', 118)],[(' ', 466), ('e', 381), ('t', 202), ('h', 136), ('d', 115)]])
#print(inpt.shape)

hmm = {}
counter = 1

for r in range(len(inpt[:,0,0])):
  smm = 0
  for c in range(len(inpt[r,:,:])):
    smm+=int(inpt[r,c,1])

  temp = []
  for c in range(len(inpt[r,:,:])):
    temp.append((str(inpt[r,c,0]),int(inpt[r,c,1])/smm))

  if counter == len(inpt[:,0,0]):
    temp.append(-1)
  else:
    temp.append(counter)

  hmm[r] = temp
  counter += 1

```

```python
def EMatch( hmmi, letter ):
  hit = (-1,-1)
  for i in hmmi[0]:
    if letter == i[0]:
      hit=i
      break
  return hit

def EProb( hmm, instr ):
  L = len( instr )
  k = 0
  pbs = 1.0
  for i in range( L ):
    emit = EMatch( hmm[k], instr[i] )
    pbs *= emit[1]
    k = hmm[k][1]
  return pbs
```
