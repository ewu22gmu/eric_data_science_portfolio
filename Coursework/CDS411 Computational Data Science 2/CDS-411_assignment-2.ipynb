## Problem 5

Using the Monte Carlo method, create 10,000 random data points.  Find all of those that are within a distance of 1 from any point on that spiral.  Return the number of points that are close to this spiral.

```python
import matplotlib.pyplot as plt

mat = np.zeros((1000,1000))

for i in range( 5000 ):

    r = i / 25
    theta = i / 200

    x = r * np.cos(theta)
    y = r * np.sin(theta)

    x = int(x * 2 + 500)
    y = int(y * 2 + 500)

    mat[x, y] = 255

plt.imsave('spiral.png', mat)
plt.imshow(mat, cmap='gray')
plt.show()

print(mat.shape)
#print(mat[500,500])
print(mat.sum()/255/(1000**2))
```

## Problem 7
The function *np.random.normal* will create random points according to a Gaussian distribution.  The function *np.random.multivariate_normal* will create random points according to a Gaussian distribution in multiple dimensions.  Create a scatter plot of 1000 points with the following qualities $\mu = (0.4,0.5)$ and $\sigma = \begin{pmatrix} 0.3 & 0 \\ 0 & 0.2 \end{pmatrix}$.  Use marker='.'  The output should be a cloud centered on $\mu$ and the widths are the diagonal elements in $\sigma$.  Use *plt.ylim* and *plt.xlim* to force the plot to have the limits of -3 to +3 in both directions.

```python
# your code and show a scatter plot of the answer
mu = (0.4, 0.5)
sigma = np.array([0.3, 0, 0, 0.2]).reshape(2,2)

md_gauss = np.random.multivariate_normal(mu,sigma, 1000)

print(md_gauss.shape)

plt.scatter(md_gauss[:,0], md_gauss[:,1],marker='.')
plt.xlim(-3, 3); plt.ylim(-3, 3)
plt.show()
```

## Problem 10
Create 1000 random points uniformally distributed between -1 and +2. Consider these as $x$ values.
Compute two new arrays, using $$y = \exp\left( \frac{-(x-\mu)^2}{2 \sigma^2} \right)$$
For the first array ($y1$) use $\mu=0.3$ and $\sigma = 0.15$, and for the second array ($y2$) use $\mu=0.6$ and $\sigma=0.15$.  

Create a scatter plot for both distributions.
You should have two bell curves.
Find the number of points (N) in which the values of $y1$ are greater than 0.01.
Compute number of points which have a larger value in $y2$ than $y1$, then normalize by N.

```python
# your code and show a plot with both distributions
x_vec = np.random.uniform(-1.0, 2.0, 1000)
mu = np.array([0.3, 0.6])
sigma = np.array([0.15, 0.15])

y1 = np.exp(-(x_vec - mu[0])**2 / (2*sigma[0]**2))
y2 = np.exp(-(x_vec - mu[1])**2 / (2*sigma[1]**2))

N = (y1[:] > 0.01).sum()
y2_gt_y1 = (y2[:] > y1[:]).sum()

plt.scatter(x_vec, y1, marker = '.', label = 'y1')
plt.scatter(x_vec, y2, marker = '.', label = 'y2')
plt.legend()
plt.show()

print(f"The number of points that have a larger value in y2 than y1 normalized by N is {y2_gt_y1/N}")
```
