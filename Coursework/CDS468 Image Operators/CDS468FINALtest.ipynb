```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.ndimage as nd
import numpy as np
import imageio
import cv2

url = 'https://raw.githubusercontent.com/ewu22gmu/CDS468Final/refs/heads/main/Q2_horse.jpg'
amg = imageio.imread(url, mode='F').astype(np.uint8)

plt.title('Original Image')
plt.imshow(amg, cmap='gray')
plt.show()

bmg = 255 - amg
plt.title('Inverted Image')
plt.imshow(bmg, cmap='gray')
plt.show()

mask = np.zeros(amg.shape).astype(bool)
mask[100:120,80:115] = True
mask[85:100,100:130] = True

mask ^= cv2.Sobel(bmg,cv2.CV_64F,1,1,ksize=5).astype(bool)

plt.title('Mask')
plt.imshow(mask, cmap='gray')
plt.show()

cmg = nd.binary_dilation(bmg, structure = np.array([[1]]), iterations=1,mask=mask)

plt.title('Image Resulting from binary dilation')
plt.imshow(cmg, cmap='gray')
plt.show()
```

```python
import imageio
import matplotlib.pyplot as plt
import imageio
import numpy as np
import warnings
from PIL import Image
from urllib.request import urlopen

warnings.filterwarnings("ignore", category=RuntimeWarning)

url = 'https://raw.githubusercontent.com/ewu22gmu/CDS468Final/refs/heads/main/Q5_dog.jpg'
adata = imageio.imread(url)

plt.title('Original Image')
plt.imshow(adata)
plt.show()

r,g,b = adata[:,:,0],adata[:,:,1],adata[:,:,2]

plt.title('R channel')
plt.imshow(r,cmap='gray')
plt.show()

plt.title('G channel')
plt.imshow(g,cmap='gray')
plt.show()

plt.title('B channel')
plt.imshow(b,cmap='gray')
plt.show()

from sklearn.decomposition import PCA
```

```python
#Create PCA objects for each channel
PCA20r = PCA(n_components=20)
PCA20r.fit(r)

PCA20g = PCA(n_components=20)
PCA20g.fit(g)

PCA20b = PCA(n_components=20)
PCA20b.fit(b)


print('\nExpained Variance Ratio Rchannel:', round(sum(PCA20r.explained_variance_ratio_),5))

print('\nExpained Variance Ratio Gchannel:', round(sum(PCA20g.explained_variance_ratio_),5))

print('\nExpained Variance Ratio Bchannel:', round(sum(PCA20g.explained_variance_ratio_),5))

PCA20r_filtered = PCA20r.inverse_transform(PCA20r.transform(r))
PCA20g_filtered = PCA20g.inverse_transform(PCA20g.transform(g))
PCA20b_filtered = PCA20b.inverse_transform(PCA20b.transform(b))

'''plt.imshow(PCA20r.inverse_transform(PCA20r.transform(r)),cmap='gray')
plt.title('R channel first 20 components')
plt.show()'''

PCA20 = np.stack([PCA20r_filtered,PCA20g_filtered,PCA20b_filtered],axis=2)
plt.imshow(PCA20.astype(np.uint8))
plt.title('PCA of first 20 components')
plt.show()
```

```python
#maybe borked Q5 code


import imageio
import matplotlib.pyplot as plt
import imageio
import numpy as np
import warnings
from PIL import Image
from urllib.request import urlopen

warnings.filterwarnings("ignore", category=RuntimeWarning)

url = 'https://raw.githubusercontent.com/ewu22gmu/CDS468Final/refs/heads/main/Q5_dog.jpg'
adata = imageio.imread(url)

plt.title('Original Image')
plt.imshow(adata)
plt.show()

r,g,b = adata[:,:,0],adata[:,:,1],adata[:,:,2]

plt.title('R channel')
plt.imshow(r,cmap='gray')
plt.show()

plt.title('G channel')
plt.imshow(g,cmap='gray')
plt.show()

plt.title('B channel')
plt.imshow(b,cmap='gray')
plt.show()

from sklearn.decomposition import PCA

def p(amg,n):
  def ScrambleImage(mgdata):
    sdata = mgdata + 0
    np.random.shuffle(sdata)
    dists = np.sqrt(((mgdata[0]-sdata).sum(1)))
    seedrow = (dists==0).nonzero()[0]
    return sdata, seedrow

  def Project(data):
    """data in rows returns matrix: projected data in rows"""
    cv = np.cov(data.transpose())
    evl, evc = np.linalg.eig(cv)
    ndata = np.dot(data, evc)
    return ndata

  sdata, seedrow = ScrambleImage(amg)
  ndata = Project(sdata)


  def Unscramble(sdata, seedrow, ndata):
    V,H = sdata.shape
    udata = np.zeros((V,H))
    udata[0] = sdata[seedrow] + 0
    unused = list(range(V))
    unused.remove(seedrow)
    nndata = ndata + 0
    k = seedrow
    for i in range( 1, V ):
        dist = np.sqrt(((nndata[k]-nndata[unused])**2).sum(1))
        ag = dist.argsort()
        k = unused[ag[0]]
        udata[i] = sdata[k]
        unused.remove( k )
    return udata
  udata = Unscramble(sdata, seedrow, ndata[:,:n])
  return udata


#Preform a PCA using the first 20 components. Show the resulting image.

PCA20r = p(r,20)
PCA20g = p(g,20)
PCA20b = p(b,20)

plt.title('R channel first 20 components')
plt.imshow(PCA20r,cmap='gray')
plt.show()

plt.title('G channel first 20 components')
plt.imshow(PCA20g,cmap='gray')
plt.show()

plt.title('B channel first 20 components')
plt.imshow(PCA20b,cmap='gray')
plt.show()

PCA20 = np.stack([PCA20r,PCA20g,PCA20b],axis=2)

plt.title('First 20 Components Image')
plt.imshow(PCA20.astype(np.uint8))
plt.show()

#Preform a PCA using the first 100 components. Show the resulting image.

n=100
PCA100r = p(r,n)
PCA100g = p(g,n)
PCA100b = p(b,n)

print(PCA100r[1,:])

plt.title('R channel first 100 components')
plt.imshow(PCA100r,cmap='gray')
plt.show()

plt.title('G channel first 100 components')
plt.imshow(PCA100g,cmap='gray')
plt.show()

plt.title('B channel first 100 components')
plt.imshow(PCA100b,cmap='gray')
plt.show()

PCA100 = np.stack([PCA100r,PCA100g,PCA100b],axis=2)
plt.title('First 100 Components Image')
plt.imshow(PCA100.astype(np.uint8))
plt.show()
```

```python
import matplotlib.pyplot as plt
import imageio
import cv2


url = 'https://raw.githubusercontent.com/ewu22gmu/CDS468Final/refs/heads/main/Q10_tiger.jpg'
amg = imageio.imread(url).astype(np.uint8)

plt.imshow(amg)
plt.title('Original Image')
plt.show()

#apply a (3,3) gaussian blur
bmg = cv2.GaussianBlur(amg,(3,3),0)

'''plt.imshow(bmg)
plt.title('Blurred Image')
plt.show()'''

print('The gaussian blur was not applied' if np.array_equal(amg,bmg) else 'The gaussian blur was applied')
```

```python
#cv2.Sobel dx=1, dy=0
sob_x = cv2.Sobel(bmg,cv2.CV_64F,1,0,ksize=5)

plt.imshow(sob_x)
plt.title('Sobel dx=1, dy=0')
plt.show()
```

```python
#cv2.Sobel dx=0, dy=1
sob_y = cv2.Sobel(bmg,cv2.CV_64F,0,1,ksize=5)

plt.imshow(sob_y)
plt.title('Sobel dx=0, dy=1')
plt.show()
```

```python
#cv2.Sobel dx=1, dy=1
sob_xy = cv2.Sobel(bmg,cv2.CV_64F,1,1,ksize=5)

plt.imshow(sob_xy)
plt.title('Sobel dx=1, dy=1')
plt.show()
```

```python
#cv2.Canny
t_L, t_U = (0, 0)

cnny = cv2.Canny(bmg, t_L, t_U)

plt.imshow(cnny, cmap='gray')
plt.title(f'Canny lower threshold = {t_L}, upper threshold = {t_U}')
plt.show()
```
