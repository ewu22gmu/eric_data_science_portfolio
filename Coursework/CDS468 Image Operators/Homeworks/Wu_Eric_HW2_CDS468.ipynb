```python
#Q1 Write a python program to list all the odd numbers from 1 to 100
print([x for x in range(1, 100+1,2)])
```

```python
#Q2 Write a python program to calculate all prime numbers between 1 and 100
num_list, primes = [x for x in range(1,100+1)], [x for x in range(2, 100+1)]

for i in num_list:
  for j in range(2, i):
    if (j != 1) & (i != j) & (i % j == 0):
      if i in primes:
        primes.remove(i)

print(primes)
```

```python
#Q3 Write a lambda function to calculate the modulo of a number with a modulus of 2
mod2 = lambda x : x % 2
print(mod2(2))
```

```python
#Q4 Starting with the lambda function in #3, write a map lambda function to calculate the iterable list x = [2, 6, 8, 10, 11, 4, 12, 7, 13, 17, 0, 3, 21]
mp_lmd = map(lambda x : x % 2, [2, 6, 8, 10, 11, 4, 12, 7, 13, 17, 0, 3, 21])
print(list(mp_lmd))
```
