## Theory

<br>

<b>The correlation of the two functions, $f(x)$ and $g(x)$, is

$c(u)=\int_{-\infty}^{\infty} f(x+u) g^{\dagger}(x) \mathrm{d} x$

where $g^\dagger(x)$ represents the complex conjugate fo $g(x)$
</b><br><br>

<b>The function $f(x)$ is shifted and multiplied by the function $g(x)$</b><br><br>

<b>The integration sums the multiplied values</b><br><br>

<b>A correlation in two-dimensions is quite similar in nature</b><br><br>

<b>$c(u, v)= \int_{-\infty} ^\infty \int_{-\infty} ^\infty f(x+u,y+v) g^\dagger (x, y) ~dx ~dy$</b><br><br>

<b>The operator notation for a correlation is $\otimes$ and is written as,

${\bf c}[\vec u] = {\bf a}[\vec x] \otimes {\bf b}[\vec x]$
</b><br><br>

<b>The output, ${\bf c}[\vec u]$, is denoted in a new space by $\vec u \in {\bf U}$</b><br><br>

<b>The correlation of a function with itself is called an auto-correlation</b><br><br>

<b>This is a useful procedure for finding presence of regular patters in a signal</b><br><br>

<b>The correlation of two different functions is also called a cross-correlation</b><br><br>

<br><br><br><br>

### Computations in Fourier Space

<br>

<b>For large two-dimensional arrays the computations of a correlation using the above equation can be slow</b><br><br>

<b>If, for example, an image was $N \times N$ pixels then the number of mult-adds (multiplications and additions) goes as $N^4$</b><br><br>

<b>A more efficient method of computing the correlation is to employ Fourier transforms</b><br><br>

<b>Consider the Fourier transform of a correlation,

$C(\alpha) = \int_{-\infty} ^\infty c(u) e^{ -\imath u \alpha } ~du$
</b><br><br>

<b>From above, this becomes

$C(\alpha) = \int_{-\infty} ^\infty  \int_{-\infty} ^\infty f(x+u) g^\dagger (x) ~ \exp \left[ -\imath u \alpha \right] ~ dx ~du$
</b><br><br>

<b>Using the substitution $z=x+u$ the equation becomes,

$C(\alpha) = \int_{-\infty} ^\infty  \int_{-\infty} ^\infty f(z) g^\dagger (x) ~ e^{ -\imath \alpha (z-x)} ~ dx ~dz$
</b><br><br>

<b>This can be rewritten as,

$C(\alpha) = \int_{-\infty}^\infty f(z) e^{-\imath \alpha z}~dz
    \int_{-\infty} ^\infty  g^\dagger (x) ~ e^{ \imath \alpha x} ~ dx
    = F(\alpha) G^\dagger (\alpha)$

where $F(\alpha)$ is the Fourier transform of $f(x)$ and $G(\alpha)$ is the Fourier transform of $g(x)$
</b><br><br>

<b>This equation computes the Fourier transform of the correlation and so an inverse transform needs to be applied to obtain the correlation</b><br><br>

<br><br>

<b>In operator notation this is,

${\bf c}[\vec \alpha] = \mathfrak{F}^{-1} \left( (\mathfrak{F} {\bf a}[\vec x]) \times ( \mathfrak{F} {\bf b}[\vec x] )^\dagger \right) = {\bf a}[\vec x] \otimes {\bf b}[\vec x]$
</b><br><br>

<b>The digital correlation for a signal of length $N$ requires $N^2$ mathematical operations</b><br><br>

<b>The digital Fourier transform for a signal of length $N$ requires $N \log_2 (N)$ operations</b><br><br>

T<b>he correlation using the Fourier methods requires three Fourier transforms and the elemental multiplication of two signals</b><br><br>

<b>Thus, the number of operations is $3 N \log_2 (N) + N$</b><br><br>

<b>Consider a case in which an image is $1024 \times 1024$</b><br><br>

<b>Then $N$ would be larger than $10^6$ and the original correlation method would require $10^{12}$ operations</b><br><br>

<b>The Fourier method, though, would only require $6.4 \times 10^7$ operations</b><br><br>

<b>For large arrays the computational savings is several orders of magnitude, and very easy to employ</b><br><br>

<br><br><br><br>

### Method Based on Fourier Transforms

<br>

<b>The <b>Correlate1D</b> function, which performs a one dimensional correlation using Fourier transforms</b><br><br>

<b>This receives two vectors of the same length and computes the correlation</b><br><br>

<b>The equivalent function for two dimensions is <b> Correlate2D</b><br><br>

<b>The only difference in the two functions besides the name is that the latter uses ft.fft2d and ft.ifft2d</b><br><br>

<b>The function <b>Correlate2DF</b> is used in cases where the input and filter are already in Fourier space</b><br><br>

<b>In some cases an image is to be correlated with several filters and so it is prudent to compute the Fourier transform of the input just once and feed it to this function for each filter</b><br><br>

<b>Both the data and the filter are already in Fourier space, and <b> Correlate2DF</b> can be used to finish the correlation computation between the inputs and filter</b><br><br>

<br><br>

```python
import scipy.fft as ft
def Correlate1D(A, B):
    a = ft.fft(A)
    b = ft.fft(B)
    c = a * b.conjugate( )
    C = ft.ifft(c)
    C = ft.fftshift(C)
    return C

def Correlate2D(A, B):
    a = ft.fft2d(A)
    b = ft.fft2d(B)
    c = a * b.conjugate( )
    C = ft.ifft2(c)
    C = ft.fftshift(C)
    return C
```

### Example - Geometric Shapes

<b>The first example is the correlation of a simple shapes</b><br><br>

<br><br>

```python
import numpy as np
import scipy.fftpack as ft

def Correlate2D(A, B):
    a = ft.fft2(A)
    b = ft.fft2(B)
    c = a * b.conjugate( )
    C = ft.ifft2(c)
    C = ft.fftshift(C)
    return C

def Circle(size, loc,rad):
    """frame, center, radius size is (v,h) of size of array loc is (v,h) of circle location
    rad is integer of radius returns array with a solid circle"""
    b1,b2 = np.indices(size)
    b1,b2 = b1-loc[0], b2-loc[1]
    mask = b1*b1 + b2*b2
    mask = (mask <= rad*rad).astype(int)
    return mask

adata = np.zeros((512,512))
adata[100:130,80:120] = 1
adata[300:350,100:210] = 1
adata[80:210,300:400] = 1
adata += Circle((512,512), (150,180), 40)
adata += Circle((512,512), (350,300), 60)

plt.title('Original Image')
plt.imshow(adata.real, cmap='gray')
plt.show()

bdata = Circle((512,512), (256,256), 60)
bdata -= Circle((512,512), (256,256), 55)

plt.title('An Annular Ring Image')
plt.imshow(bdata.real, cmap='gray')
plt.show()

cdata = Correlate2D(adata, bdata)

plt.title('Correlation of the Original and Annular Ring Images')
plt.imshow(-cdata.real, cmap='gray')
plt.show()
```

### Example - Boat Isolation

<br>

<b>In this example individual boats will be isolated and labeled</b><br><br>

<b>The main problem in this task is that the boats and the dock touch and that the range of pixel intensities of the boats are beyond the range of the dock</b><br><br>

<b>So, it is not possible to separate the dock from the boats using a threshold</b><br><br>

<b>The construction of the dock is known and in this task the image is considered on only one scale and orientation</b><br><br>

<b>A correlation is used to locate the dock, and then it can be subtracted from the original image</b><br><br>

<b>Isolation of the boats is easy once the dock is removed</b><br><br>

<br><br>

<b>There are four steps in this process</b><br><br>

<b>The first is to load the original image and create the dock mask</b><br><br>

<b>The second step is to locate the mask in the original image</b><br><br>

<b>The third step is to subtract the dock</b><br><br>

<b>The final step is to clean the remaining components of the image to create isolated boats</b><br><br>

<br><br>
<div><img src="https://github.com/joefoxva1/CDS468/blob/master/Lectures10_11_12/boats1.png?raw=true" width="500"/img></div>
<br><br>

<b>The original image is a color image extracted from a satellite image which is designated as ${\bf a}[\vec x]$</b><br><br>

<b>This image is converted to gray scale as,

${\bf b}[\vec x] = \mathcal{L}_L {\bf a}[\vec x]$
</b><br><br>

<b>The dock dimensions are known (perhaps from previous satellite images or other intelligence)</b><br><br>

<b>The mask of the dock is created from simple rectangles</b><br><br>

<b>The function BuildDock creates this mask</b><br><br>

<b>The list crn contains the horizontal left edge of each rectangle</b><br><br>

<b>The loop creates rectangles that are 84 pixels high and 4 wide</b><br><br>

<b>This dock needs to be centered in the frame before being used in a correlation and so nd.center_of_mass finds the center of mass and nd.shift centers the dock in the frame</b><br><br>

<br><br>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import imageio
import scipy.ndimage as nd
import cv2

def BuildDock():
    dock = np.zeros((189,545))
    dock[39:46,0:508] = 1
    dock[0:39,9:14] = 1
    dock[46:84,0:5] = 1
    crn = [30,66,100,136,172,206,242,281,315, 349,384,420,456,492]
    for c in crn:
        dock[0:84,c:c+4] = 1
    V,H = dock.shape
    v,h = nd.center_of_mass(dock)
    dock = nd.shift(dock + 0., (V/2-v, H/2-h))
    return dock

url = 'https://raw.githubusercontent.com/joefoxva1/CDS468/master/Lectures10_11_12/boats1.png'
orig = imageio.imread(url, as_gray=True).astype(np.uint8)
orig = cv2.resize(orig, dsize=(545,189))  #  resize the image

plt.title('Original Image Grayscale')
plt.imshow(orig, cmap='gray')
plt.show()

dock = BuildDock()

plt.title('Dock Image')
plt.imshow(-dock, cmap='gray')
plt.show()
```

<br><br>

<b>The gray scale boat image is represented by ${\bf b}[\vec x]$ and the dock mask is represented by ${\bf m}[\vec x]$</b><br><br>

<b>The intent is to locate the position of the dock in the original image and then to subtract the dock which will separate the boats from other pixels in the same intensity range</b><br><br>

<b>The process of locating a boat is performed by correlating the dock mask with the image</b><br><br>

<b>The peak of the correlation is directly related to the location of the center of the dock in the original image</b><br><br>

<b>As long as the dock is centered in the mask then the amount of shift required to align the mask to the dock in the image is the distance between the center of the frame and the location of the correlation peak</b><br><br>

<b>The location of the peak is extracted by the $A_\vee$ operator which is applied to the correlation of the image with the dock mask as in,

$\vec v = A_\vee \left( {\bf b}[\vec x] \otimes {\bf m}[\vec x] \right)$
</b><br><br>

<br><br>

```python
import scipy.fftpack as ft

def Correlate2D(A, B):
    a = ft.fft2(A)
    b = ft.fft2(B)
    c = a * b.conjugate( )
    C = ft.ifft2(c)
    C = ft.fftshift(C)
    return C

def LocateDock(orig, dock):
    corr = Correlate2D(orig, dock)
    V,H = corr.shape
    v,h = divmod(abs(corr).argmax(), H)
    return v,h

vh = LocateDock(orig, dock)
print('\nLocation of the center of the dock', vh)
```

<br><br>

<b>The code below will show the dock positioned over the image</b><br><br>

<b>The method by which this image is created begins with the definition of the vector $\vec c$ as the center of the image frame</b><br><br>

<b>The process is defined as

${\bf d}[\vec x] = {\bf b}[\vec x] + 300 \Gamma_{>0.1} D_{\vec v - \vec c} {\bf m}[\vec x]$
</b><br><br>

<b>The mask is shifted by $\vec v - \vec c$ which positions the dock</b><br><br>

<b>A small threshold is applied as the shift function does install small artifacts if the input images has integer values</b><br><br>

<b>The result is multiplied by a large number and then added to the original</b><br><br>

<b>This image is merely to determine if the dock is correctly placed and will not be used in further computations</b><br><br>

<br><br>

```python
def Overlay(orig, dock, vh):
    v,h = vh
    V,H = dock.shape
    ndock = nd.shift(dock, (v-V/2, h-H/2))
    ndock = ndock > 0.1
    return orig + 300 * ndock

olay = Overlay(orig, dock, vh)

plt.title('Dock on original Image')
plt.imshow(-olay, cmap='gray')
plt.show()
```

<br><br><br><br>

<b>The third step in the process is to subtract the dock from the original image, and the fourth step is to isolate the boats. Both steps can be performed by,

${\bf f}[\vec x] = \Gamma_{>140} \left( {\bf b}[\vec x] - 200 ~ D_{\vec v - \vec c} \lhd  \Gamma_{>0.1} {\bf m}[\vec x] \right)$
</b><br><br>

<b>Once again a small threshold is applied to the dock before the dilation process is applied as in $\lhd \Gamma_{>0.1} {\bf m}[\vec x]$</b><br><br>

<b>This is to ensure that all of the background pixels are exactly 0</b><br><br>

<b>If there is even a small bit error then the dilation will not work properly</b><br><br>

<b>This result is shifted by $D_{\vec v - \vec c}$ so that dock is centered in the frame and then multiplied by a large value</b><br><br>

<b>This is subtracted from the original thus making the dock pixels much darker than anything else in the image</b><br><br>

<b>The threshold $\Gamma_{>140}$ passes only those pixels that are brighter than 140 which excludes all of the ocean pixels and dock pixels</b><br><br>

<b>The surviving pixels belong only to the boats</b><br><br>

<b>The process is contained within the SubtractDock function</b><br><br>

<br><br>

```python
def SubtractDock(orig, dock, vh):
    v,h = vh
    V,H = dock.shape
    dock2 = nd.binary_dilation(dock > 0.1, iterations=1)
    ndock = nd.shift(dock2+0., (v-V/2, h-H/2))
    ndock = (orig - 200 * ndock)>140
    return ndock

isoboats = SubtractDock(orig, dock, vh)

plt.title('Original Image')
plt.imshow(orig, cmap='gray')
plt.show()

plt.title('Dock Subtracted from Original Image')
plt.imshow(~isoboats, cmap='gray')
plt.show()
```

<br><br><br><br>

<b>Not all of the boats in the image are contiguous segments</b><br><br>

<b>Some of the boats have separate segments due to vastly different pixel values for the boat deck and cabin canopy</b><br><br>

<b>The final step is to remove spurious pixels and to unite boats that might be shown in more than one contiguous segment</b><br><br>

<b>This step relies on the idea that all of the boats are facing vertically in the image</b><br><br>

<b>The dilation process is modified so that there is a dilation in one direction that is different than the dilation in the other direction</b><br><br>

<b>In this case, the desire is to dilate in the vertical direction which would connect separated boat segments</b><br><br>

<b>This dilation must be large enough to cross the gaps between segments but small enough to not bridge over the dock</b><br><br>

<br><br>

<b>The process starts with the definition of ${\bf P}$ which is a matrix that is $7 \times 1$ and all of the values are 1</b><br><br>

<b>The description is,

${\bf g}[\vec x] = \mathcal{I} \lhd_{\bf P} {\bf f}[\vec x]$
</b><br><br>

<b>The dilation $\lhd_{\bf P}$ will connect segments that are within 7 pixels of each other in the vertical direction, thus uniting the boat segments</b><br><br>

<b>The final step is to label the components</b><br><br>

<b>These final steps are performed in the IDboats function</b><br><br>

<b>The result below displays each isolated segment with a different pixel intensity</b><br><br>>

## Composite Filtering

<br>

<b>Correlation filters are somewhat useful but in many real applications the target can have several presentations</b><br><br>

<b>In the previous example, the search was limited such that the dock could have only a single orientation and magnification</b><br><br>

<b>A more realistic case would consider any orientation and a range of scales</b><br><br>

<b>It would be possible to create a filter for every possible orientation/scale combination but that would be a prohibitive computation</b><br><br>

<b>A better solution is to use composite filtering which creates a filter from several training images</b><br><br>

<b>Thus, one single filter could identify the location of a target over a range of orientations and scales</b><br><br>

<b>Once the filter is built the computation time to identify the dock is the same as that of a single correlation</b><br><br>

<br><br><br><br>

## SDF and MACE

<br>

<b>SDF is the synthetic discriminant function</b><br><br>

<b>MACE is the minimum average correlation energy filter</b><br><br>

<b>The composite filter theory starts with the desire to create a filter, $\vec h$, that solves,

$\vec c = {\bf X}\vec h$

where $c_i = 1 ~~ \forall i$  and the columns of ${\bf X}$ are the target data vectors
</b><br><br>

<b>Extrapolation of this system to two-dimensional data is straightforward as it is possible to vertorize the image, placing all of its pixels in a single column vector</b><br><br>

<b>If ${\bf X}$ were square then the solution would simply require the computation of the inverse of ${\bf X}$</b><br><br>

<b>However, when used for images ${\bf X}$ is generally rectangular</b><br><br>

<b>Therefore, the pseudo-inverse is employed and the filter is computed by

$\vec h = {\bf X}\left( {\bf X}^\dagger {\bf X} \right)^{-1} \vec c$
</b><br><br>

<b>The center pixel of the correlation between the filter $\vec h$ and any of the inputs $\vec x_i$ will produce $c_i$</b><br><br>

<b>This value is also $\vec h \cdot \vec x$</b><br><br>

<b>Thus, $\vec c$ is called the constraint vector as it constrains a single value in the correlation, but the rest of the values are unconstrained</b><br><br>

<br><br>

<b>It is possible that correlation values other than the center have a higher magnitude</b><br><br>

<b>In these cases, the peak may not be accurately identified</b><br><br>

<b>A proposed solution is to simultaneously reduce the total correlation energy as well as constrain the center value</b><br><br>

<b>This led to the minimum average correlation energy (MACE) filter</b><br><br>

<b>The MACE is computed by,

$\hat{\vec{h}}=\mathbf{D}^{-1 / 2} \mathbf{X}\left(\hat{\mathbf{X}}^{\dagger} \mathbf{D}^{-1} \hat{\mathbf{X}}\right)^{-1} \overrightarrow{\boldsymbol{c}}$

where,

$D_{i,j} = \frac{\delta_{i,j}}{N} \sum_k | \hat x_i ^{(k)} |^2$

and $x_i ^{(k)}$ represents the i-th element of the Fourier transform of the k-th vector
</b><br><br>

<b>The MACE must be computed with the data in Fourier space and this is designated by the carets</b><br><br>

<b>The MACE filter builds filters that produce a very sharp correlation peak when correlated with a training input</b><br><br>

<b>However, if the correlation is with the filter and another vector that is not used in training the height of the correlation peak is small</b><br><br>

<b>This is the case even if the correlation input vector is mostly similar to a training vector</b><br><br>

<br><br>

## Fractional Power Filter

<br><br>

### Theory

<br>

<b>The matrix <b>D</b> is a diagonal matrix and if the power term where 0 instead of 2 then <b>D</b> would be the Identity matrix</b><br><br>

<b>The only difference between the two filters is the power term and thus, the FPF (fractional power filter) with

$D_{i,j} = \frac{\delta_{i,j}}{N} \sum_k | \hat x_i ^{(k)} |^\alpha$

where $0 \le \alpha \le 2$
</b><br><br>

<b>The operator $Q_{\vec c,\alpha} \left\{ {\bf a}_i[\vec x] \right\}$ represents the process of creating an FPF filter</b><br><br>

<b>The vector $\vec c$ is the constraint vector and is an optional argument</b><br><br>

<b>If it is not present then it is assumed that all of the elements in this vector are set to 1 which coincides with the original use of the filter</b><br><br>

<b>The $\alpha$ is the fractional power value between 0 and 2, and the images inside of the curly braces represent the training data</b><br><br>

<br><br>

<b>Thus, the SDF can be computed by

${\bf b}[\vec x] = Q_0 \left\{ {\bf a}_i[\vec x] \right\}$
</b><br><br>

<b>The FPF is constructed on the Fourier transform of the images and thus it is defined as

${\bf b}[\vec \omega] = Q_\alpha \left\{ \mathfrak{F} {\bf a}_i[\vec x] \right\}$
</b><br><br>

<b>This filter is in Fourier space and thus has the coordinates defined by $\vec \omega$</b><br><br>

<b>In some applications there is an efficiency to be gained by leaving the filter in the Fourier space instead of converting it back to the original image space</b><br><br>

<br><br>

```python
#  Fractional Power Filter
def FPF(data, c, fp):
    (N, Dim)= data.shape
    D = (np.power(abs(data), fp)).sum(0)
    D = D / N
    ndx = (abs(D) < 0.001).nonzero()[0]
    D[ndx] = 0.001 * np.sign(D[ndx] + 1e9)
    Y = data / np.sqrt(D)
    Y = Y.transpose()
    Yc = Y.conjugate().transpose()
    Q = Yc.dot(Y)
    if N == 1:
        Q = 1./Q
    else:
        Q = np.linalg.inv(Q)
    Rc = Q.dot(c)
    H = Y.dot(Rc) / np.sqrt(D)
    return H

data = np.random.ranf((3,10))  #  three input vectors of lenght 10
c = np.ones(3)     # constraint vector
h = FPF(data, c, 0)  #  apply the fractional filter; output corresponds to the three values in the
#                         constraint vector

print('\nInput Dot Product (center of a correlation result) Input 1:', round(h.dot(data[0]),2))

print('\nInput Dot Product (center of a correlation result) Input 2:', round(h.dot(data[1]),2))

print('\nInput Dot Product (center of a correlation result) Input 3:', round(h.dot(data[2]),2))
```

<br><br><br><br>

### Manipulating $\alpha$

<br>

<b>The value of $\alpha$ ranges from 0 to 2 and when applied to images the behavior is somewhat predictable</b><br><br>

<b>The value ${\bf D}_{i,i}$ from above is proportional to the magnitude of frequency $i$ in the training data</b><br><br>

<b>However, it only uses the inverse of $\bf D$</b><br><br>

<b>Thus, the filter has high magnitudes for low magnitude frequencies in the training data</b><br><br>

<b>When applied to most types of images the increase in $\alpha$ corresponds to a reliance of the edge information in the training images</b><br><br>

<b>This occurs since the low frequencies are often very large and correspond to solid shapes in the input</b><br><br>

<b>Basically, for image of objects, an increase in $\alpha$ is related to a greater sensitivity to the object's edges</b><br><br>

<b>As $\alpha$ goes from 0 to 2, the filter trades off generalization for discrimination</b><br><br>

<b>A simple example of the effect of $\alpha$ begins with an image that has various, solid geometric shapes</b><br><br>

<br><br>

<b>The creation of the filter is

${\bf h}[\vec x] = \mathfrak{F}^{-1} \mathcal{V}_{\vec w_1} Q_{\vec c,\alpha} \mathcal{C}_m \left\{ \mathcal{V}_{\vec w_2} \mathfrak{F}{\bf b}[\vec x] \right\}$
</b><br><br>

<b>In the code below the original data is loaded as ${\bf a}[\vec x]$ and prepares a scaled version by

${\bf b}[\vec x] = 1 - \frac{1}{255} \times \mathcal{L}_L {\bf a}[\vec x]$
</b><br><br>

<b>The process $\mathcal{V}_{\vec w_2} \mathfrak{F}{\bf b}[\vec x]$ computes the Fourier transform of the image and reshapes it to conform the $\vec w_2$</b><br><br>

<b>In this case, $\vec w_2$ represents a column in the matrix with the length of $V \times H$</b><br><br>

<b>The Concatenation operator $\mathcal{C}_m$ puts all of these vectors into a matrix. In the example, there is only one image and it is placed into the matrix</b><br><br>

<b>In this case, $\alpha=0$</b><br><br>

<b>The final two operators, $ \mathfrak{F}^{-1} \mathcal{V}_{\vec w_1}$ are performed in line 11 where $\vec w_1$ represents the original image frame size</b><br><br>

<b>When using a correlation, the first filter will generalize well</b><br><br>

<b>The input shapes do not have to match those in the training image precisely</b><br><br>

<b>However, the correlation peak may be difficult to detect particularly if there is any noise in the input</b><br><br>

<b>The second filter will discriminate well</b><br><br>

<b>It can better distinguish between to similar shapes because it is more sensitive to shape (the outline or edges of the objects)</b><br><br>

<br><br>

```python
#  Fractional Power Filter
import imageio
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc as sm
import scipy.fftpack as ft

def FPF(data, c, fp):
    # data is the incoming data... in rows : It is converted to columns here c is the constraint vector
    (N,Dim)= data.shape
    #X = transpose( data )# put vectors into columns

    # Compute D.  Best done in rows
    D = (np.power(abs(data), fp)).sum(0)
    D = D / N

    # it is possible that some values of D are 0 which will bomb out later
    ndx = (abs(D) < 0.001).nonzero()[0]
    D[ndx] = 0.001 * np.sign(D[ndx]+1e9)

    # Y is the modified X.  Also more efficient to compute from original data
    Y = data / np.sqrt(D)
    Y = Y.transpose()

    # compute Q
    Yc = Y.conjugate().transpose()
    Q = Yc.dot(Y)  # inner product

    if N == 1:  # only 1 training vector
        Q = 1./Q
    else:
        Q = np.linalg.inv(Q)

    Rc = Q.dot(c)
    H = Y.dot(Rc) / np.sqrt(D)
    # to test:  sum(conjugate(H[:,0]) * data[:,any] should equal c[any]
    return H

url = 'https://raw.githubusercontent.com/joefoxva1/CDS468/master/Lectures10_11_12/geos.png'
bdata = imageio.imread(url, mode='F').astype(np.uint8)

plt.title('Original Image')
plt.imshow(bdata, cmap='gray')
plt.show()

V,H =  bdata.shape
Xt = np.zeros((1,V*H), complex)
Xt[0] = ft.fft2(bdata).ravel()
cst = np.ones(1)

alpha = 0.3
ffilt = FPF(Xt, cst, alpha)
filt = ft.ifft2(ffilt.reshape(V,H))

plt.title('Alpha = 0.3 Image')
plt.imshow(filt.real, cmap='gray')
plt.show()

alpha = 0.3
ffilt = FPF(Xt, cst, alpha)
filt = abs(ft.ifft2(ffilt.reshape(V,H)))

print(filt.real[400,400])
plt.title('Absolute value of Alpha = 0.3 Image')
plt.imshow(-filt.real, cmap='gray')
plt.show()
```

## Restrictions of Correlations

<br>

<b>The FPF is a good tool but it does have limitations which are common among correlation filters</b><br><br>

<b>Correlation filters are inherently first order</b><br><br>

<b>This means that the filter responds to matches with the input vector</b><br><br>

<b>While it is possible to have the filter respond to a particular shape (such as the circle example) it is very difficult to build a filter that will respond to a circle that does not have a rectangle nearby</b><br><br>

<b>In this latter case the problem is second order and would need a stronger tool</b><br><br>

<b>Correlation filters are translation invariant</b><br><br>

<b>In the case of the circle filter, it does not matter where the circle is located in the frame</b><br><br>

<b>The correlation spike will be located wherever the target is located</b><br><br>

<b>However, there are variances that cause performance decay in the filters</b><br><br>

<b>Consider the case of searching for a rectangle of a known ratio</b><br><br>

<b>A filter could be constructed to identify this rectangle, but if the query image has a rectangle of a different scale or a different rotation then the filter may not provide a good response</b><br><br>

<b>The correlation filters are not rotation invariant or scale invariant</b><br><br>

<b>There are a few solutions that have been proposed</b><br><br>

<b>One option is to construct an FPF with several rotations and scales of the target</b><br><br>

<b>The advantage of this transformation is that the scale and rotation now become linear shifts in the new space</b><br><br>

<b>Since the filters are translation invariant, filters built in the polar space are rotation and scale invariant</b><br><br>

<b>The disadvantage of this conversion is that the center of the object must be known, and this information may not be available in some applications</b><br><br>

## Summary

<br>

<b>An image correlation is equivalent to the alteration of the image frequencies in accordance with frequencies from a second image</b><br><br>

<b>This second image in small kernel correlations has only a few elements</b><br><br>

<b>Kernels for enhancing edges or smoothing the input image are easily deployed</b><br><br>

<b>The computations for large kernel correlations use Fourier transforms which reduces the computation time by orders of magnitude</b><br><br>

<b>A common implementation is to create a filter from an image that contains a target centered in the frame</b><br><br>

<b>The correlation of this with an input that contains the target will produced a correlation surface with a large spike at the location of the target</b><br><br>

<b>A composite filter is created from multiple images and has a control factor that manages the trade-off between generalization and discrimination</b><br><br>

<b>This process can also be trained to ignore or reject some targets</b><br><br>

<b>Dual composite filters can quantify the changing parameters in a set of images even if that parameter is not well defined</b><br><br>
