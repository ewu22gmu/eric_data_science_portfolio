## Eigenimages

<br>

<b>Eigenimages are created in a process similar to PCA, but with a modification. As a reminder, the traditional PCA process is:

1. Warp images to a single grid
2. Compute the covariance matrix of the images
3. Compute the eigenimages from the covariance matrix
4. Downselect the eigenimages to use

</b><br><br>

<b>Step 2, though, is potentially an untenable step</b><br><br>

<b>The size of the covariance matrix can be prohibitively large </b><br><br>

<b>If the images were $ 256 \times 256$, which is a small image, then the covariance matrix would be $65,536 \times 65,536$</b><br><br>

<b>Using single precision floats, this array would consume 16 GB</b><br><br>

<b>Thus, this method needs to be modified in order to calculate the eigenimages</b><br><br>

<br><br>

<b>Formally, the process starts with a set of images, ${\bf a}_i[\vec x]; i=1,\ldots,P$</b><br><br>

<b>Every image is warped to a single grid, $G$</b><br><br>

<b>$\left\{ {\bf b}[\vec x] \right\} = \left\{ W_G {\bf a}[\vec x] \right\}$</b><br><br>

<b>The average images is computed</b><br><br>

<b>${\bf \bar b}[\vec x] = \frac{1}{N} \sum  \left\{ {\bf b}[\vec x] \right\}$</b><br><br>

<b>and subtracted from each image</b><br><br>

<b>$\left\{ {\bf c}[\vec x] \right\} = \left\{ {\bf b}[\vec x] - {\bf \bar b}[\vec x] \right\}$</b><br><br>

<b>The notation $\left\{ {\bf b}[\vec x] - {\bf \bar b}[\vec x] \right\}$ is a set</b><br><br>

<br><br>

<b>There are several images in $\{ {\bf b}[\vec x]\}$ but only a single image ${\bf \bar b}[\vec x]$</b><br><br>

<b>Since the latter resides inside of the brackets, it is subracted from each image</b><br><br>

<b>The output is a set of images which are the original images with the average image subtracted</b><br><br>

<b>In the original theory, the covariance matrix would be computed from these images</b><br><br>

<b>From this the eigenimages are computed by,

$\left.\{\mathbf{d}[\vec{x}]\}=\frac{1}{\sqrt{\lambda_{j}}}\left\{\begin{array}{c}\emptyset \\ \Downarrow_{\lambda_{j}<\gamma}\end{array}\right\} \mathcal{E} V\left\{\mathbf{c}_{[} \vec{x}\right]\right\}$

</b><br><br>

<b>that matrix would be extremely large</b><br><br>

<br><br><br><br>

### Large Covariance Matrix

<br>

<b>The solution is to consider an alternate but similar problem.  The eigen--equation is,

${\bf C}\vec x_i = \lambda_i \vec x_i$

</b><br><br>

<b>and that the covariance matrix can be expressed by ${\bf C} = {\bf X}{\bf X}^T$ where ${\bf X}$ is the matrix that contains the original data in columns.  Therefore,</b><br><br>

${\bf X}{\bf X}^T \vec v_i = \lambda_i \vec v_i$

<b>For image data this matrix is prohibitively large.  Consider a new eigen-equation,</b><br><br>

${\bf L}\vec w_i = \lambda_i \vec w_i$

<b>where ${\bf L} = {\bf X}^T {\bf X}$.  Left multiply the equation by ${\bf X}^T$ to get,

${\bf X}^T{\bf X}{\bf X}^T \vec v_i = \lambda_i {\bf X}^T \vec v_i$

therefore,

${\bf L}{\bf X}^T \vec v_i. = \lambda_i {\bf X}^T \vec v_i$

</b><br><br>
    
<b>The matrix ${\bf L}$ is $P \times P$ where $P$ is the number of images and since $P<<N$ the size of ${\bf L}$ is much smaller than the size of ${\bf C}$</b><br><br>

<b>The solution is to then solve for the eigenvectors and eigenvalue of ${\bf L}$ and then compute $\vec v$'s from $\vec w$'s</b><br><br>

<br><br><br><br>

###  PCA for visualization: Hand-written digits

<br>

</b><br><br>To demonstrate PCA on a more interesting problem, let's consider one piece of the optical character recognition problem: the identification of hand-written digits</b><br><br>

</b><br><br>In the wild, this problem involves both locating and identifying characters in an image</b><br><br>

</b><br><br>Here we'll take a shortcut and use Scikit-Learn's set of pre-formatted digits, which is built into the library</b><br><br>

<br><br>

```python
#  load the handwritten digits
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits

digits = load_digits()
print('\nDigits shape:', digits.data.shape)
```

<br><br>

<b>The data consists of 8Ã—8 pixel images, meaning that they are 64-dimensional</b><br><br>

<b>To gain some intuition into the relationships between these points, we can use PCA to project them to a more manageable number of dimensions, say two</b><br><br>

<br><br>

```python
from sklearn.decomposition import PCA

pca = PCA(2)  # project from 64 to 2 dimensions
projected = pca.fit_transform(digits.data)
print('\nDigits shape:', digits.data.shape)
print('\nDigits projected shape:', projected.shape)
```

```python
#  Plot the two principle components
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral',10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar()
plt.show()
```

<br><br>

<b>Recall what these components mean: the full data is a 64-dimensional point cloud, and these points are the projection of each data point along the directions with the largest variance</b><br><br>

<b>Essentially, we have found the optimal stretch and rotation in 64-dimensional space that allows us to see the layout of the digits in two dimensions, and have done this in an unsupervised mannerâ€”that is, without reference to the labels</b><br><br>

<br><br><br><br>

### Choosing the number of components

<b>A vital part of using PCA in practice is the ability to estimate how many components are needed to describe the data</b><br><br>

<b>This can be determined by looking at the cumulative explained variance ratio as a function of the number of components</b><br><br>

<br><br>

```python
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

<br><br>

<b>This curve quantifies how much of the total, 64-dimensional variance is contained within the first N components</b><br><br>

<b>For example, we see that with the digits the first 10 components contain approximately 75% of the variance, while you need around 50 components to describe close to 100% of the variance</b><br><br>

<b>Here we see that our two-dimensional projection loses a lot of information (as measured by the explained variance) and that we'd need about 20 components to retain 90% of the variance</b><br><br>

<b>Looking at this plot for a high-dimensional dataset can help you understand the level of redundancy present in multiple observations</b><br><br>

<br><br><br><br>

### PCA as Noise Filtering

<br>

<b>PCA can also be used as a filtering approach for noisy data</b><br><br>

<b>The idea is this: any components with variance much larger than the effect of the noise should be relatively unaffected by the noise</b><br><br>

<b>So if you reconstruct the data using just the largest subset of principal components, you should be preferentially keeping the signal and throwing out the noise</b><br><br>

<b>Let's see how this looks with the digits data. First we will plot several of the input noise-free data</b><br><br>

<br><br>

```python
#  show some digits
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                             subplot_kw={'xticks':[], 'yticks':[]},
                             gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap='binary', interpolation='nearest',
                  clim=(0, 16))
plot_digits(digits.data)
```

```python
#  add some noise
np.random.seed(42)
noisy = np.random.normal(digits.data, 4)
plot_digits(noisy)
```

<br><br>

<b>It's clear by eye that the images are noisy, and contain spurious pixels. Let's train a PCA on the noisy data, requesting that the projection preserve 50% of the variance</b><br><br>

<br><br><br><br>

```python
#  how many components constitue 50% of the variance in the noisy set?
pca = PCA(0.5).fit(noisy)
print('\nNumber of components:', pca.n_components_)

#  check this
pca2 = PCA(n_components=12)
pca2.fit(noisy)
print('\nExpained Variance Ratio:', round(sum(pca2.explained_variance_ratio_),2))
```

<br><br>

<b>Here 50% of the variance amounts to 12 principal components of the noisy digits</b><br><br>

<b>Now we compute these components, and then use the inverse of the transform to reconstruct the filtered digits</b><br><br>

<br><br>

```python
#  use the 12 components
components = pca.transform(noisy)
filtered = pca.inverse_transform(components)
plot_digits(filtered)
```

<br><br><br><br>

###  Summary

<b>This signal preserving/noise filtering property makes PCA a very useful feature selection routineâ€”for example, rather than training a classifier on very high-dimensional data, you might instead train the classifier on the lower-dimensional representation, which will automatically serve to filter out random noise in the inputs</b><br><br>
