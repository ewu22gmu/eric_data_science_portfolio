## Loops

<b>There are two types of loops in Python, for and while, <br>
1. For loops iterate over a given sequence
2. While loops repeat as long as a certain boolean condition is met</b>

<br><br>

```python
# for loop example
sum = 0 
for i in range (100):
    sum += 1
print('\nLoop total is:', sum)
```

```python
# while loop
i = 0
sum = 0
while i < 100: 
    sum += i
    i += 1
print('\nWhile total is:', sum)
print('\nLoop total is:', i)
```

```python
# break out of a loop
## end the for loop if i is equal to 3
for i in range (10):
    if i==3:
        break
    print(i)
```

```python
# continue: Skip the rest of the loop body, but do not end the loop
## skip print statement if i equal to 3
for i in range (10):
    if i==3:
        continue 
    print(i)
```

## Map and Lambda Functions

<b>Combining map and lambda can replace a simple loop<br><br>
lambda creates a simple function<br>
&nbsp;&nbsp;&nbsp;&nbsp;A lambda operator can have any number of arguments but can have only one expression<br>
&nbsp;&nbsp;&nbsp;&nbsp;It cannot contain any statements and returns a function object which can be assigned to any variable<br>

Map provides the iterations<br>
&nbsp;&nbsp;&nbsp;&nbsp;map functions expect a function object and any number of iterables, such as list, dictionary, etc.<br>
&nbsp;&nbsp;&nbsp;&nbsp;It executes the function object for each element in the sequence and returns a list of the elements modified by the function object</b>

<br><br>

```python
# lambda example
# typical function
def add(x, y): 
    return x + y
  
# Call the function
print('\nFunction output:', add(2, 3))
```

```python
add = lambda x, y : x + y
print('\nlambda output:', add(2, 3))
```

```python
# map example
def multiply2(x):
  return x * 2

y = map(multiply2, [1, 2, 3, 4])
print('\nmap output', list(y))
```

```python
#  putting lambda and map together
y = map(lambda x : x*2, [1, 2, 3, 4]) 
print('\nlambda and map output', list(y))
```

## Useful Image Functions

<br>

<b>argmax – used to provide the location of the maximum; returns a single value of the location<br>
argmin - used to provide the location of the minimum; returns a single value of the location<br>

Example:<br>

Image has 653 elements in a row with 519 rows<br>
total elements  =  338,907<br><br>

Find the vertical-horizontal location<br>
divmod(image.argmax(),653)  =  (40, 326)<br><br>

Find the total number of the maximum valued pixels<br>
(image==image.max()).sum()  =  28</b>

<br><br>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import imageio
import scipy.ndimage as nd


url = 'https://raw.githubusercontent.com/joefoxva1/CDS468/master/Lecture2_3_4/bird_js.jpg'
adata = imageio.imread(url)

plt.title('Original Image')
plt.imshow(adata)
plt.show()

print('\nImage shape:', adata.shape)

print('\nLocation of the first largest pixel value:', adata.argmax())

print('\nMaximum pixel value:', adata.max())

print('\nMinimum pixel value:', adata.min())

print('\nVertical-horizontal location of the first largest pixel value:', divmod(adata.argmax(),255))

print('\nTotal number of the maximum valued pixels:', (adata==adata.max()).sum())
```

### Jacobian Matrix - J

<br>

Jacobian carries information about the local behavior of F and can be thought of as a local expansion/distortion factor for volumes<br><br>

The Jacobian matrix is defined as the matrix of all its first-order partial derivatives<br><br>

Jacobian matrix represents the differential of f at every point where f is differentiable<br><br>

The Jacobian is the best linear approximation of the change of f in a neighborhood of x<br><br>

$J_{i j}=\frac{\partial}{\partial x_{j}} f(x)$<br><br><br><br>

<div> <img src="https://github.com/joefoxva1/GGS_590_DL/blob/master/Picures/LA_Jacobian.png?raw=true" width="500"/img></div>

### Hessian Matrix - H

<br>

The Hessian matrix is defined as the second-order partial derivatives of a scalar-valued function, or scalar field<br><br>

Hessian matrix represents the second derivative of f at every point where f is differentiable<br><br>

It describes the local curvature of a function of many variables<br><br>

$H_{i j}=\frac{\partial^{2} f(x)}{\partial x_{i} \partial x_{j}}$

## scipy

<br>

A scientific computing package for Python<br><br>

SciPy greatly extends the functionality of the NumPy routines<br><br>

```python
#  example
%matplotlib inline
import matplotlib.pyplot as plt
import imageio
import scipy.ndimage as nd

url = 'https://raw.githubusercontent.com/joefoxva1/CDS468/master/Lecture2_3_4/bird_js.jpg'
amg = imageio.imread(url)

plt.title('Original Image')
plt.imshow(amg)
plt.show()

print('\nImage shape:', amg.shape)
print('\nData type:', amg.dtype)

#Rearrange color channels
bmg = amg.transpose((2,0,1))
print('\nImage reshape:', bmg.shape)
```

```python
#  Rotation and Shift
#  Rotation is measured by an angle and a shift is by a scaler
from scipy import ndimage

print('\nImage rotation example')
fmg = ndimage.rotate(amg, 20, reshape=False, cval=255)  #rotate image by 20 degrees and pixels the not defined, set to white (cval=255)
plt.imshow(fmg)
plt.show()

print('\nImage shift example')
emg = ndimage.shift(amg, (25,10,0)) # shift image by 25 pixels in the vertical direction, 10 pixels in the horizontal direction and no color shift
plt.imshow(emg)
plt.show()

# Center of Mass, to calculate the center of mass of the bird image
print('\nCenter of Mass of the bird:', nd.center_of_mass(amg))
```
